```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
df = pd.read_csv("../datasets/raw/IndianWeatherRepository.csv")
print("Dataset Loaded Successfully")
```

```python
multivariate_features = [
    "temperature_celsius",
    "humidity",
    "pressure_mb",
    "wind_kph",
    "precip_mm",
    "cloud",
    "visibility_km",
    "uv_index",
    "air_quality_PM2.5",
    "air_quality_PM10"
]

df[multivariate_features].head()
```

```python
pairplot_features = [
    "temperature_celsius",
    "humidity",
    "pressure_mb",
    "precip_mm",
    "air_quality_PM2.5"
]

sns.pairplot(
    df[pairplot_features].sample(3000, random_state=42)
)

plt.show()


print("""
Pairplot reveals relationships
among major climate variables.

Useful for:
- Clustering
- Feature Engineering
- Model Development
""")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="air_quality_PM2.5",
    y="air_quality_PM10",
    hue="air_quality_us-epa-index",
    alpha=0.4
)
plt.title("PM2.5 vs PM10 by EPA Index")
plt.show()


print("""
Pollution indicators move together.
High EPA index observations
represent severe pollution events.
Useful for anomaly detection.
""")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="temperature_celsius",
    y="air_quality_PM2.5",
    alpha=0.4
)
plt.title("Temperature vs PM2.5")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="humidity",
    y="cloud",
    size="precip_mm",
    alpha=0.5
)
plt.title("Humidity + Cloud vs Rainfall")
plt.show()


print("""
Rainfall tends to occur when:
- Humidity increases
- Cloud cover increases
Strong candidate predictors
for rainfall risk modeling.
""")
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="temperature_celsius",
    y="uv_index",
    size="feels_like_celsius",
    alpha=0.5
)
plt.title("Temperature + UV vs Feels Like Temperature")
plt.show()


print("""
Heatwave conditions are associated with:
- High temperature
- High UV index
- High feels-like temperature
Important for heatwave prediction.
""")
```

```python
cluster_features = [
    "temperature_celsius",
    "humidity",
    "pressure_mb",
    "precip_mm",
    "visibility_km",
    "uv_index",
    "air_quality_PM2.5",
    "air_quality_PM10"
]

df[cluster_features].describe()
```

```python
anomaly_features = [
    "temperature_celsius",
    "precip_mm",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "air_quality_Carbon_Monoxide"
]
df[anomaly_features].describe()


print("""
Extreme values exist in:
- Temperature
- Rainfall
- PM2.5
- PM10
- Carbon Monoxide
These variables are strong
candidates for Isolation Forest.
""")
```

```python
rainfall_features = [
    "humidity",
    "cloud",
    "pressure_mb",
    "wind_kph",
    "temperature_celsius",
    "precip_mm"
]
df[rainfall_features].corr()
```

```python
heatwave_features = [
    "temperature_celsius",
    "feels_like_celsius",
    "humidity",
    "uv_index",
    "cloud"
]

df[heatwave_features].corr()
```
